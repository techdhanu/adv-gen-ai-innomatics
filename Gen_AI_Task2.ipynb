```python
from transformers import pipeline

llm = pipeline("text-generation", model="google/flan-t5-base")

result = llm("Explain LangChain in simple terms", max_new_tokens=100)
print(result[0]['generated_text'])
```

```python
def prompt_template(topic):
    return f"Explain {topic} in simple terms with an example"

prompt = prompt_template("Machine Learning")
result = llm(prompt, max_new_tokens=100)

print(result[0]['generated_text'])
```

```python
def simple_chain(topic):
    step1 = f"What is {topic}?"
    step2 = f"Give a real-world example of {topic}"

    res1 = llm(step1, max_new_tokens=50)[0]['generated_text']
    res2 = llm(step2, max_new_tokens=50)[0]['generated_text']

    return res1 + "\n\n" + res2

print(simple_chain("Artificial Intelligence"))
```

```python
memory = []

def chat(input_text):
    memory.append(input_text)
    context = " ".join(memory)

    response = llm(context, max_new_tokens=100)[0]['generated_text']
    return response

print(chat("Hi"))
print(chat("What is AI?"))
print(chat("Give example"))
```

```python
def calculator(expression):
    return eval(expression)

print("Calculation:", calculator("25 * 4"))
```

```python
def agent(query):
    if "calculate" in query.lower():
        expr = query.split("calculate")[-1]
        return calculator(expr)
    else:
        return llm(query, max_new_tokens=100)[0]['generated_text']

print(agent("calculate 45 + 5"))
print(agent("Explain deep learning"))
```

```python
def chatbot(user_input):
    return llm(user_input, max_new_tokens=100)[0]['generated_text']

print(chatbot("What is LangChain?"))
```
