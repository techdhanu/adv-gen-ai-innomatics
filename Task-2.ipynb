1. Unique Words in a Sentence

```python
sentence = input("Enter sentence: ")
words = sentence.split()
unique_words = set(words)

print(f"Unique words count: {len(unique_words)}")
print(f"Unique words: {unique_words}")
```

2.Highest Salary from Employee Data


```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

highest_paid_employee = max(employees, key=employees.get)
highest_salary = employees[highest_paid_employee]

print(f"Highest Salary: {highest_paid_employee} - {highest_salary}")
```

3.Find Maximum and Minimum Values

```python
numbers = [45, 22, 89, 10, 66]
maximum = numbers[0]
minimum = numbers[0]
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num
print(f"List: {numbers}")
print(f"Max: {maximum}")
print(f"Min: {minimum}")
```

4.Count Products Above a Price Threshold

```python
prices = [450, 1200, 899, 1500, 300]
count = 0
for price in prices:
    if price > 1000:
        count += 1

print(f"Products above 1000: {count}")
```

5.Calculate Attendance Percentage

```python
attendance = ["P", "P", "A", "P", "P"]

present_count = attendance.count("P")
total_days = len(attendance)
attendance_percentage = (present_count / total_days) * 100

print(f"Attendance Percentage: {attendance_percentage}")
```
