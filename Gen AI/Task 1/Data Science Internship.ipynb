```python
##Task 4: Token Analytics
import re

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'http\S+|www\.\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'[^\w\s]', '', text)
    words = text.split()
    exceptions = {"no", "not"}
    return [w for w in words if len(w) > 2 or w in exceptions]

samples = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []
for s in samples:
    tokens = preprocess_text(s)
    results.append({
        "original": s,
        "cleaned": tokens,
        "noise_removed": len(s.split()) - len(tokens),
        "meaningful_retained": len(tokens)
    })

most_noise = max(results, key=lambda r: r["noise_removed"])
most_meaningful = max(results, key=lambda r: r["meaningful_retained"])
print("Sentence with most noise removed:")
print("Original:", most_noise["original"])
print("Cleaned:", " ".join(most_noise["cleaned"]))
print("Noise Removed:", most_noise["noise_removed"])
print("\nSentence retaining most meaningful tokens:")
print("Original:", most_meaningful["original"])
print("Cleaned:", " ".join(most_meaningful["cleaned"]))
print("Meaningful Tokens:", most_meaningful["meaningful_retained"])
```

```python
##Task 4: Token Analytics
import re

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'http\S+|www\.\S+', '', text)       # remove URLs
    text = re.sub(r'\S+@\S+', '', text)                # remove emails
    text = re.sub(r'\d+', '', text)                    # remove numbers
    text = re.sub(r'(.)\1{2,}', r'\1', text)           # collapse repeated chars
    text = re.sub(r'[^\w\s]', '', text)                # remove emojis/punct
    words = text.split()
    exceptions = {"no", "not"}
    filtered_words = [w for w in words if len(w) > 2 or w in exceptions]
    return filtered_words

samples = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for s in samples:
    tokens = preprocess_text(s)
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = sum(len(t) for t in tokens) / total_tokens if total_tokens > 0 else 0
    
    results.append({
        "original": s,
        "tokens": tokens,
        "sentence": " ".join(tokens),
        "total_tokens": total_tokens,
        "unique_tokens": unique_tokens,
        "avg_length": round(avg_length, 2),
        "noise_removed": len(s.split()) - total_tokens
    })

for r in results:
    print("Original Text:", r["original"])
    print("Cleaned Tokens:", r["tokens"])
    print("Cleaned Sentence:", r["sentence"])
    print("Total Tokens:", r["total_tokens"])
    print("Unique Tokens:", r["unique_tokens"])
    print("Average Token Length:", r["avg_length"])
    print("-" * 60)

most_noise = max(results, key=lambda r: r["noise_removed"])
most_meaningful = max(results, key=lambda r: r["total_tokens"])

print("\nAnalysis:")
print("Sentence with most noise removed:")
print("Original:", most_noise["original"])
print("Cleaned:", most_noise["sentence"])
print("Noise Removed:", most_noise["noise_removed"])

print("\nSentence retaining most meaningful tokens:")
print("Original:", most_meaningful["original"])
print("Cleaned:", most_meaningful["sentence"])
print("Meaningful Tokens:", most_meaningful["total_tokens"])
```

```python
##Task 5: Frequency Analysis
import re
from collections import Counter

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'http\S+|www\.\S+', '', text)       # remove URLs
    text = re.sub(r'\S+@\S+', '', text)                # remove emails
    text = re.sub(r'\d+', '', text)                    # remove numbers
    text = re.sub(r'(.)\1{2,}', r'\1', text)           # collapse repeated chars
    text = re.sub(r'[^\w\s]', '', text)                # remove emojis/punct
    words = text.split()
    exceptions = {"no", "not"}
    return [w for w in words if len(w) > 2 or w in exceptions]

samples = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

all_tokens = []
for s in samples:
    all_tokens.extend(preprocess_text(s))
counter = Counter(all_tokens)

top_10 = counter.most_common(10)
least_5 = counter.most_common()[:-6:-1]  # last 5

print("Top 10 Most Frequent Words:")
for word, freq in top_10:
    print(f"{word}: {freq}")

print("\nTop 5 Least Frequent Words:")
for word, freq in least_5:
    print(f"{word}: {freq}")
```

```python
##Task 6: Build Full Pipeline
import re

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'http\S+|www\.\S+', '', text)       # remove URLs
    text = re.sub(r'\S+@\S+', '', text)                # remove emails
    text = re.sub(r'\d+', '', text)                    # remove numbers
    text = re.sub(r'(.)\1{2,}', r'\1', text)           # collapse repeated chars
    text = re.sub(r'[^\w\s]', '', text)                # remove emojis/punct
    words = text.split()
    exceptions = {"no", "not"}
    return [w for w in words if len(w) > 2 or w in exceptions]

def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []
    
    for text in text_list:
        tokens = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(" ".join(tokens))
    
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

samples = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
output = full_pipeline(samples)

print("Pipeline Output:")
print(output)
```
