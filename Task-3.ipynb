1. Employee Performance Bonus Eligibility


```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

max_score = max(employees.values())
top_performers = [name for name, score in employees.items() if score == max_score]
print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {max_score})")
```

2.  Search Query Keyword Analysis



```python
import string

search_query = "Buy mobile phone buy phone online"
query_lower = search_query.lower()
query_clean = query_lower.translate(str.maketrans('', '', string.punctuation))
words = query_clean.split()
word_count = {}
for word in words:
    word_count[word] = word_count.get(word, 0) + 1

result = {word: count for word, count in word_count.items() if count > 1}

print(result)

```

3. Sensor Data Validation

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = [(index, value) for index, value in enumerate(sensor_readings) if value % 2 == 0]

print(f"Valid Sensor Readings (Hour, Value): {valid_readings}")

```

4.Email Domain Usage Analysis

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}
for email in emails:
    domain = email.split('@')[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

total_users = len(emails)
for domain, count in domain_count.items():
    percentage = (count / total_users) * 100
    print(f"{domain}: {percentage:.0f}%")
```

5.Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

average_sales = sum(sales) / len(sales)
threshold = average_sales * 1.30

for index, sale in enumerate(sales):
    if sale > threshold:
        print(f"Day {index + 1}: {sale}")
```

6. Duplicate User ID Detection

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

id_count = {}
for user_id in user_ids:
    id_count[user_id] = id_count.get(user_id, 0) + 1

for user_id, count in id_count.items():
    if count > 1:
        print(f"{user_id} → {count} times")
```
