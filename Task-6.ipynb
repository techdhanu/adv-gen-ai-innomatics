1.Smart Parking Lot Management System

```python
def smart_parking_system(capacity, vehicle_logs):
    parked = 0
    peak = 0

    for log in vehicle_logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1

        if parked > peak:
            peak = parked

    status = "Available"
    if parked > capacity:
        status = "Over Capacity"
    elif parked == capacity:
        status = "Full"

    print("Currently Parked Vehicles:", parked)
    print("Parking Status:", status)

# Sample Test
smart_parking_system(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

2.Online Food Delivery Time Estimator

```python
def delivery_time_estimator(distance, traffic, weather):
    base_time = distance * 5

    if traffic == "High":
        base_time += 15
    elif traffic == "Medium":
        base_time += 8

    if weather == "Rainy":
        base_time += 10
    elif weather == "Storm":
        base_time += 20

    print("Estimated Delivery Time:", base_time, "minutes")

# Sample Test
delivery_time_estimator(8, "High", "Rainy")
```

3.Movie Theatre Seat Occupancy Analyzer

```python
def theatre_occupancy(total_seats, booked_seats):
    booked = len(booked_seats)
    occupancy = (booked / total_seats) * 100

    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"

    print("Occupancy:", int(occupancy), "%")
    print("Show Status:", status)

# Sample Test
theatre_occupancy(200, [1]*150)
```

4.Cloud Server Load Classification System

```python
def server_load_classifier(cpu_readings):
    avg_cpu = sum(cpu_readings) / len(cpu_readings)

    if avg_cpu < 50:
        status = "Normal"
    elif 50 <= avg_cpu <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print("Average CPU Load:", int(avg_cpu), "%")
    print("Server Status:", status)

# Sample Test
server_load_classifier([45, 60, 70, 85, 90])
```

5.Smart Classroom Resource Usage Monitor

```python
def classroom_resource_monitor(resource_usage):
    overused = []

    for resource, hours in resource_usage.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")

# Sample Test
classroom_resource_monitor({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})
```

6.Online Event Registration Capacity Controller

```python
def event_registration_controller(capacity, registrations):
    if registrations > capacity:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlisted = 0
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)

# Sample Test
event_registration_controller(100, 105)
```
