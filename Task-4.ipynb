1.Mobile Recharge Validation System

```python
def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: ₹"))

        if amount < 50:
            print("Minimum recharge amount is ₹50. Try again.")
        elif amount not in valid_plans:
            print("Invalid plan. Available plans:", valid_plans)
        else:
            print("Recharge Successful!")
            break

validate_recharge()
```

2.Inventory Reorder Alert System


```python
def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(product, ": Reorder Alert")
        else:
            print(product, ": Stock OK")


inventory = {
    "Rice": 20,
    "Sugar": 10,
    "Oil": 5,
    "Wheat": 25
}

check_inventory(inventory)
```

3.Student Result Processing System

```python
def process_result(marks):
    total = 0

    for mark in marks:
        total += mark

    average = total / len(marks)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"


marks_list = [60, 70, 45, 55]
result = process_result(marks_list)
print("Result:", result)
```

4.Cab Fare Estimator with Retry Option



```python
def calculate_fare(distance, peak_hour):
    base_fare = 50
    rate_per_km = 12

    fare = base_fare + (distance * rate_per_km)

    if peak_hour.lower() == "yes":
        fare = fare + (fare * 0.25)

    return fare


while True:
    try:
        distance = float(input("Enter distance in km: "))
        peak = input("Is it peak hour? (yes/no): ")

        total_fare = calculate_fare(distance, peak)
        print("Total fare:", total_fare)

        retry = input("Do you want to calculate again? (yes/no): ")
        if retry.lower() != "yes":
            break
    except ValueError:
        print("Invalid input. Please enter correct values.")
```

5. Employee Attendance Eligibility Checker


```python
def check_eligibility(attendance):
    total_days = len(attendance)
    present_days = 0

    for day in attendance:
        if day == "P":
            present_days += 1

    percentage = (present_days / total_days) * 100

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


attendance_list = ["P", "P", "A", "P", "P", "A", "P"]
status = check_eligibility(attendance_list)
print("Attendance Status:", status)
```

6. Password Strength Checker

```python
def check_password(password):
    if len(password) < 8:
        return "Weak Password"

    has_digit = False
    has_special = False

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in "@#$":
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"


password_input = input("Enter password: ")
print(check_password(password_input))
```
