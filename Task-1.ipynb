2. Pass / Fail Analyzer

```python
marks = [45, 78, 56, 89, 90]
pass_count = 0
fail_count = 0
for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1
print(f"Total Students: {len(marks)}")
print(f"Pass Students: {pass_count}")
print(f"Fail Students: {fail_count}")
```

3. Simple Data Cleaner

```python
names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []
for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)

print("Original Names:", names)
print("Cleaned Names:", cleaned_names)
```

4. Message Length Analyzer


```python
messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    length = len(message)
    print(f"Message: '{message}' - Length: {length}", end="")

    if length > 10:
        print(" [LONG MESSAGE]")
    else:
        print()
```

5. Error Message Detector

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0
for log in logs:
    if log == "ERROR":
        error_count += 1

print(f"Total Logs: {len(logs)}")
print(f"Total Errors: {error_count}")
```
